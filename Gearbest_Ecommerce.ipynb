```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

### Mobiles

```python
mobile_final_df = pd.DataFrame()

for j in range(1,103):
    url = "https://www.gearbest.ma/collections/cell-phones-accessories/page/{}/".format(j)
    web = requests.get(url)
    soup = BeautifulSoup(web.text,"html.parser")

    price = []
    model = []
    reviews = []
    thumps_up = []

    b = soup.find_all('div',class_="price_for_grid redbrightcolor floatleft fontbold")
    for i in b:
        price.append(i.text.strip())

    c = soup.find_all('h3')
    for i in c:
        if i:
            model.append(i.text)
        else:
            model.append(np.nan)

    d = soup.find_all('span',class_="rh_opacity_5 font80 ml10")
    for i in d:
        if i:
            reviews.append(i.text)
        else:
            reviews.append(np.nan)

    e = soup.find_all('span',class_="thumbscount")
    for i in e:
        if i:
            thumps_up.append(i.text)
        else:
            thumps_up.append(np.nan)

    # print(len(price))
    # print(len(model))
    # print(len(reviews))
    # print(len(thumps_up))

    di = {'Price':price,'Model_name':model,'Reviews':reviews,'Liked':thumps_up}

    df = pd.DataFrame(di)

    mobile_final_df = pd.concat([mobile_final_df,df],ignore_index=True)
```

```python
mobile_final_df['Category'] = 'Mobile'
```

```python
mobile_final_df
```
