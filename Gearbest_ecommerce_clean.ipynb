### Price column

```python
df.Price
```

```python
df[df.Price.isna()]
```

```python
df[df.Price.isna()].count()
```

```python
df = df.dropna(subset=['Price'])
```

```python
df.info()
```

```python
df['Price(in $)'] = df.Price.str.split('$').str.get(0)
```

```python
df['Price(in $)'] = df['Price(in $)'].str.replace(',','')
```

```python
temp = df[df['Price(in $)'].str.contains('â')]
```

```python
# df['Price(in $)'] = df['Price(in $)'].astype(float)
```

```python
temp_price_values = temp.Price.str.split('â').str.get(0)
temp_price_index = temp.Price.str.split('â').str.get(0).index
```

```python
df.loc[temp_price_index, 'Price(in $)'] = temp_price_values
```

```python
df[df['Price(in $)'].str.contains('â')]
```

```python
df['Price(in $)'] = df['Price(in $)'].str.replace(',','')
```

```python
df[df['Price(in $)'].str.contains(',')]
```

```python
df['Price(in $)'] = df['Price(in $)'].astype(float)
```

```python
df.info()
```

```python
df.drop(columns=['Price'],inplace=True)
```

```python
df['Price(in $)'] = df['Price(in $)'].round(1)
```
