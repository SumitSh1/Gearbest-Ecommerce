```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Gearbest_ecommerce_clean.xlsx')
```

### Bivariate Analysis

#### 1. Category - Price analysis

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.barplot(data=df,x='Category',y='Price(in $)',hue='Category')
plt.xticks(rotation=90)

plt.subplot(1,2,2)
sns.boxplot(data=df,x='Category',y='Price(in $)',hue='Category')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12,4))
sns.stripplot(data=df,x='Category',y='Price(in $)',jitter=True, hue='Category')
plt.xticks(rotation=90)
plt.show()
```

#### 2. Category - Reviews analysis

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df,x='Category',y='Reviews',hue='Category')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12,4))
sns.stripplot(data=df,x='Category',y='Reviews',hue='Category')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12,4))
sns.violinplot(data=df,x='Category',y='Reviews',hue='Category')
plt.xticks(rotation=90)
plt.show()
```

#### 3. Category - Likes analysis

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='Category',y='Liked',hue='Category')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12,4))
sns.stripplot(data=df,x='Category',y='Liked',hue='Category')
plt.xticks(rotation=90)
plt.show()
```

### Conclusions
1. In almost all categories, the maximum number of reviews posted under products belonging to that specific product categories ranges from 1 to 3. A surge in reviews upto 10 can be observed. Possibly couldbe due to positive and negative aspects of the product. This shows that customers are usually not interested in posting reviews.
2. Maximum number of likes ranges from 0 to 4 across all the product categories with 0 being the highest hich certainly means that most of the customers are not interested in giving likes to the products. Products under few categories have received good number of likes which basically means the product might be of good quality.
3. Maximum number of products fall under the product category consumer electronics followed by mobiles.
4. The price range for most of the products across all categories ranges from 0 to $500.
5. The average price of the product category Health & Beauty is the highest compared to other product categories.
6. 3 products under product category Mobile has received highest number of reviews ranging upto 70.
